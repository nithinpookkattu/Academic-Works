Linear Regression Model with Machine Learning Approach

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

#importing data from .csv file
data = pd.read_csv(r'C:\Users\Nithin\Desktop\House Data\Bangalore.csv')
#print(data)

decision_1 = input('Do you need to select X & Y Values from the dataset? (Y/N): ')

if decision_1 == 'Y':
    
    print('\n')
    
    print(data.head())
    
    print('\n')
    
    x_value = input('Choose the X Variable from the dataset: ')

    print('\n')

    y_value = input('Choose the Y Variable from the dataset: ')
    
    #converting the data into list
    x = data[x_value].values
    #print(x)
    y = data[y_value].values
    #print(y)

else:
    #converting the data into list
    x = data['X'].values
    #print(x)
    y = data['Y'].values
    #print(y)

print('\n')

decision_2 = input('Do you need reduce the X Values (Y/N): ')

if decision_2 == 'Y':
    
    print('\n')
    
    x_red = input('By what multipier do you need to reduce X?: ')
    x_red = int(x_red)
    x = x/x_red
    #print(x)
else:
    x = x
    
print('\n')
    
decision_3 = input('Do you need reduce the Y Values (Y/N): ')

if decision_3 == 'Y':
    
    print('\n')
    
    y_red = input('By what multipier do you need to reduce Y?: ')
    y_red = int(y_red)
    y = y/y_red
    #print(y)
else:
    y = y

print('\n')    
    
#getting user input for Learning Rate
LR = input('Learning Rate in %: ')
LR = float(LR)/100
#print(LR)

print('\n')

#getting user input for number of epoch(s) required
NOE = input('Number of epoch required: ')
NOE = int(NOE)
#print(NOE)

n = len(x)
#print(n)

Th_0 = 0
Th_1 = 0
sse_list = list()
Th_0_lst = list()
Th_1_lst = list()
diff_lst = list()

for i in range(1, NOE+1):
    for i in range(n):
        y_pre = (Th_1*x[i])+Th_0
        #print(y_pre)
        error = y_pre - y[i]
        #print(error)
        Th_0 = Th_0 - (LR*error)
        Th_1 = Th_1 - (LR*error*x[i])

    sse_total = 0
    #diff_lst = list()
    Th_0_lst.append(Th_0)    
    Th_1_lst.append(Th_1)
    
    for i in range(n):
        y_p = (Th_1*x[i]) + Th_0
        diff = (y_p - y[i])
        sse = diff**2
        sse_total = sse_total + sse
        
    diff_lst.append(diff)
    sse_list.append(sse_total)

print('\n')
NOE = str(NOE)
#sum of squared error after every epoch
print('Sum of squared errors after '+NOE+ ' Epoch(s) is/are',sse_list)

print('\n')

comparing SSE after each epoch
plt.plot(sse_list, label='SSE')
plt.xlabel('Epoch')
plt.ylabel('SSE Value')
```
