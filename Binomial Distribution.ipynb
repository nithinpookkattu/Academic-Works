Consider a game where you are trying to make a ball in a basket.

You are given 10 shots and you know that you have an 80% chance of making a given shot. Each shot is an independent event.

n = Total Shots
p = Chance of success

```python
from scipy.stats import binom
data_binom = binom.rvs(n=10,p=0.8,loc=0,size=1000)
#print(data_binom)

import seaborn as sns
ax = sns.distplot(data_binom,kde=True,color='red',hist_kws={"linewidth": 20,'alpha':1})
ax.set(xlabel='Binomial', ylabel='Frequency')

# Assign and print probability of 8 or less successes
prob1 = binom.cdf(k=8, n=10, p=0.8)
print(prob1)
```
